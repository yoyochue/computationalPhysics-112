```python


import numpy as np
import matplotlib.pyplot as plt
import solver_ivp as solver


def f(x,v,arg):
    omega0=arg[0]
    gamma=arg[1]

    return [v,-omega0**2*x-2*gamma*v]

#x,v,t=solve_ivp(f,[0,40], [1,0], "RK2", 0.1,(1.0,1.0,0.1))
A=0.01
omega0=1
gamma=1
omega1=np.sqrt(abs(omega0**2-gamma**2))
x2,v2,t2=solver.solve_ivp(f,[0,40], [0,np.sqrt(omega1)*A],"RK4", 0.01,(omega0,gamma))
u=omega1*x2
w=gamma*x2+v2

#plt.plot(t,x,label="RK2")
#plt.plot(t1,x1,label="euler")
plt.plot(t2,x2,label="RK4")
plt.xlabel("t(s)")
plt.ylabel("x(m)")
plt.title(f"x-t $\gamma$={gamma}\nDamped Harmonic Oscillator ")
plt.legend()
plt.subplots()
plt.plot(t2,v2,label="RK4")
plt.xlabel("t(s)")
plt.ylabel("v(m/s)")
plt.title(f"v-t $\gamma$={gamma}\nDamped Harmonic Oscillator ")
plt.legend()
plt.subplots()
plt.plot(u,w,label="RK4")
plt.xlabel("u")
plt.ylabel("w")
plt.title(f"Phase space $\gamma$={gamma}\nDamped Harmonic Oscillator ")
plt.legend()
plt.axis("equal")
plt.show()
```

```python


import numpy as np
import matplotlib.pyplot as plt
import solver_ivp as solver


def f(x,v,arg,t):
    omega0=arg[0]
    lamb=arg[1]
    omegaf=arg[2]
    F0=arg[3]

    return [v,F0*np.cos(omegaf*t)-omega0**2*x-lamb*v]

#x,v,t=solve_ivp(f,[0,40], [1,0], "RK2", 0.1,(1.0,1.0,0.1))
A=0.01
omega0=1
gamma=1
omega1=np.sqrt(abs(omega0**2-gamma**2))
x2,v2,t2=solver.solve_ivp(f,[0,40], [0,omega1*A],"RK4", 0.01,(omega0,gamma))
u=omega1*x2
w=gamma*x2+v2

#plt.plot(t,x,label="RK2")
#plt.plot(t1,x1,label="euler")
plt.plot(t2,x2,label="RK4")
plt.xlabel("t(s)")
plt.ylabel("x(m)")
plt.title(f"x-t $\gamma$={gamma}\nDamped Harmonic Oscillator ")
plt.legend()
plt.subplots()
plt.plot(t2,v2,label="RK4")
plt.xlabel("t(s)")
plt.ylabel("v(m/s)")
plt.title(f"v-t $\gamma$={gamma}\nDamped Harmonic Oscillator ")
plt.legend()
plt.subplots()
plt.plot(u,w,label="RK4")
plt.xlabel("u")
plt.ylabel("w")
plt.title(f"Phase space $\gamma$={gamma}\nDamped Harmonic Oscillator ")
plt.legend()
plt.axis("equal")
plt.show()
```
